```python
from sklearn.model_selection import train_test_split
```

```python
features= goodreadsbooks.drop(columns=['new_average_rating'])               
target= goodreadsbooks['new_average_rating']
```

```python
target
```

```python
features
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.15, random_state = 0)
```

```python
from sklearn.model_selection import GridSearchCV

n_estimators = [30,60]
max_depth = [20,30]
```

```python
grid = {'n_estimators':n_estimators,
       'max_depth':max_depth}
```

```python
 from sklearn.ensemble import RandomForestClassifier #Random Forest Classifier with Grid Search CV

forest = RandomForestClassifier()
grid_search = GridSearchCV(estimator = forest, param_grid = grid, cv = 5)
grid_search.fit(X_train, y_train)
```

```python
grid_search.score(X_test,y_test) #our machine learning model has a score of 93%
```

```python
xxxtest=pd.DataFrame(grid_search.predict(X_test))
xxxtest[0].value_counts()
```
