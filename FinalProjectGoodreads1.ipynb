```python
import re
for i in range(len(goodreadsbooks.genre)):
    r1 = re.findall(r"[:\n\"](\w+\'?w* *-*\w*\'?w* *\w*\'?w*)[\"]",str(goodreadsbooks.genre[i]))
    goodreadsbooks.loc[i, 'genre'] = str(r1)
    #(goodreadsbooks.genre[i])
    #print(goodreadsbooks.genre[i])
goodreadsbooks.head()
#clean the rows without trash
```

```python
goodreadsbooks["genre"] = goodreadsbooks["genre"].apply(eval)
```

```python
x = sum(goodreadsbooks.genre, [])
x #these are all the genres that we have
```

```python
from collections import Counter
c = Counter(x)
print(c) #count all the different genres 
```

```python
c.most_common(8) #the 8 genres more common
```

```python
lis = []
for x in c.most_common(8):
    lis.append(x[0])
lis
```

```python
def search(row):
    genre_new= []
    for el in row:
        if el in lis:
            genre_new.append(el)
        else:
            genre_new.append('other')
    return list(set(genre_new)) #for the others genres that are not included in these 8 most common, we have 'other'
```

```python
goodreadsbooks['genre'] = goodreadsbooks['genre'].apply(search)
```

```python
goodreadsbooks.head()
```

```python
goodreadsbooks['new_genre'] = goodreadsbooks['genre'].apply(lambda x : '|'.join(x))
```

```python
goodreadsbooks
```

```python
#goodreadsbooks['fiction']=goodreadsbooks['new_genre'].apply(lambda x: x.contains('Fiction'))
goodreadsbooks['fiction']=goodreadsbooks['new_genre'].str.contains('Fiction').astype(int)
goodreadsbooks['novel']=goodreadsbooks['new_genre'].str.contains('Novel').astype(int)
goodreadsbooks['speculative_fiction']=goodreadsbooks['new_genre'].str.contains('Speculative fiction').astype(int)
goodreadsbooks['children_literature']=goodreadsbooks['new_genre'].str.contains('Children').astype(int)
goodreadsbooks['science_fiction']=goodreadsbooks['new_genre'].str.contains('Science Fiction').astype(int)
goodreadsbooks['mystery']=goodreadsbooks['new_genre'].str.contains('Mystery').astype(int)
goodreadsbooks['suspense']=goodreadsbooks['new_genre'].str.contains('Suspense').astype(int)
goodreadsbooks['fantasy']=goodreadsbooks['new_genre'].str.contains('Fantasy').astype(int)
#a kind of dummies; it creates a new column for each of one of 8 most relevant genres of books                                                                               
goodreadsbooks
```
